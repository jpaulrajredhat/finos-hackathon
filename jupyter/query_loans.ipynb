```python
conn = connect(
    host="trino",       # Use container name if inside Docker
    port=8080,
    user="admin",
    catalog="iceberg",
    schema="single_family",
)
```

```python
# query = "SELECT * FROM loans LIMIT 10"
query = "SELECT * FROM loans"
df = pd.read_sql(query, conn)
print("Total rows:", len(df))
print(df.head())
```

```python
col_info = [(col, str(dtype)) for col, dtype in zip(df.columns, df.dtypes)]
print(tabulate(col_info, headers=["Column", "Data Type"]))
```

```python
# print(tabulate(df.head(10), headers='keys', tablefmt='pretty'))
selected_columns = ["loan_identifier", "channel", "loan_purpose","loan_age", "origination_date",]
print(tabulate(df[selected_columns].head(800), headers='keys', tablefmt='fancy_grid'))

```
